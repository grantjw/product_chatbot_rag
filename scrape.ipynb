```python
import csv
from youtube_transcript_api import YouTubeTranscriptApi

# Define the input and output file names
input_filename = 'youtube_url.csv'
output_filename = 'youtube_transcripts.csv'

# Initialize a set to hold the unique URLs and a list for all URLs
unique_urls = set()
all_urls = []

# Read the YouTube URLs from the input CSV and check for duplicates
with open(input_filename, mode='r', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        url = row[0]
        all_urls.append(url)
        unique_urls.add(url)
        
# Compare the sizes to check for duplicates
if len(all_urls) != len(unique_urls):
    print(f"Found {len(all_urls) - len(unique_urls)} duplicate URL(s).")
else:
    print("No duplicate URLs found.")

# Initialize a list to hold the data for the output CSV
output_data = []

# Process each unique URL
for url in unique_urls:
    print(f'Processing {url}')

    try:
    # Extract the video ID from the URL
        video_id = url.split('v=')[1].split('&')[0]
    
    # Attempt to fetch the transcript
    
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        transcript_text = ' '.join([item['text'] for item in transcript])
        output_data.append([url, transcript_text])
    except Exception as e:
        pass

# Write the output data to a new CSV file
with open(output_filename, mode='w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['url', 'Transcript'])  # Write the header
    writer.writerows(output_data)

print('Process completed.')
```
